**What is AIRBNB?**

Airbnb is an online service for lodging.'Airbnb' means *'AirBed and Breakfast'*.

Airbnb clients have the opportunity to give for rent their housing partially or fully to different travelers. The site provides a platform for establishing contact between the owner and the guest.

## <b> Since 2008, guests and hosts have used Airbnb to expand on traveling possibilities and present a more unique, personalized way of experiencing the world. #  **EXPLORATORY DATA ANALYSIS**

# 1.What can we learn about different hosts and areas?

```python
hosts_areas = new_df.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
top_hosts=hosts_areas.sort_values(by='calculated_host_listings_count', ascending=False).head(5)
top_hosts
```

```python
# Visualizing the hosts with most listings.
plt.rcParams['figure.figsize'] = (10,6)
host_name = top_hosts['host_name']
host_lisitng = top_hosts['calculated_host_listings_count']
plt.bar(host_name,host_lisitng)
plt.title('Hosts with most listings in NYC',{'fontsize':18})
plt.xlabel('Host Names',{'fontsize':18})
plt.ylabel('Number of host listings',{'fontsize':18})
plt.show()
```

**Observations**:From the above result,we can conclude that the name of the hosts with the most listings is **Sonder(NYC)** who has *327 listings* in **Manhattan**


# 2.What can we learn from predictions? (ex: locations, prices, reviews, etc)

```python
areas_reviews = new_df.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
areas_reviews
```

```python
# Visualising the number of reviews in each neighbourhood group through bar plot using seaborn
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'neighbourhood_group', y = 'number_of_reviews', data =areas_reviews, color = "salmon")
plt.show()
```

Above Bar plot stated that Queens, Manhattan have majority of reviews which implies that people are liking these neighbourhoods. 

```python
price_area = new_df.groupby(['price'])['number_of_reviews'].max().reset_index()
price_area.head(5)
```

```python
area = price_area['price']
price = price_area['number_of_reviews']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.scatter(area, price)
 
plt.xlabel("Price")
plt.ylabel("Number of Review")
plt.title("Price vs Number of Reviews")
plt.show()
```

**Observations**:From the above result most of the people prefer to stay  in place where the prices is low.

# 3.Which hosts are the busiest and why?

```python
#Maximum numbers of reviews as people are booking frequently in those airbnb.
busiest_hosts = new_df.groupby(['host_name','host_id','room_type'])['number_of_reviews'].max().reset_index()
busiest_hosts = busiest_hosts.sort_values(by='number_of_reviews', ascending=False).head(10)
busiest_hosts
```

```python
name = busiest_hosts['host_name']
reviews = busiest_hosts['number_of_reviews']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(name, reviews, color ='brown',
        width = 0.4)
 
plt.xlabel("Name of the Host")
plt.ylabel("Number of Reviews")
plt.title("Busiest Hosts")
plt.show()
```

```python
# Plotting histogram for room types to look for which are most preferred room types.
plt.rcParams['figure.figsize'] = (8,5)
hp = sns.histplot(df['room_type'], color= 'red')
hp.set_xlabel('Room type') 
hp.set_ylabel('Number of listings')
plt.show()
```

The above mentioned hosts are busiest because they host Entire home/apt and private room which are the most preferred room types according to above histogram plot.

# 4.Is there any noticeable difference of traffic among different areas and what could be the reason for it?

```python
traffic_areas = new_df.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_areas = traffic_areas.sort_values(by='minimum_nights', ascending=False)
traffic_areas
```

```python
room_type = traffic_areas['room_type']
stayed = traffic_areas['minimum_nights']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(room_type, stayed, color ='maroon',
        width = 0.4)
 
plt.xlabel("Room Type")
plt.ylabel("Minimum number of nights stayed")
plt.title("Traffic Areas")
plt.show()
```

From the Above Analysis We can say that People are preferring Entire home/apt or Private room which are present in Manhattan, Brooklyn, Queens and people are preferring listings which are less in price.
